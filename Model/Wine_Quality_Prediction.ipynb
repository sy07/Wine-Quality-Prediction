```python
# splitting data into X and y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']]
y = df['quality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30) 
```

#Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_squared_log_error,r2_score
```

```python
print("Mean Absolute Error:", mean_absolute_error(y_test, lr_pred))
print("Mean Squared Error:", mean_squared_error(y_test, lr_pred))
print("Mean Squared Log Error:", mean_squared_log_error(y_test, lr_pred))
print("R2 Score", r2_score(y_test, lr_pred))
```

#Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lgr = LogisticRegression(solver='newton-cg')
```

```python
lgr.fit(X_train, y_train)
```

```python
lgr_pred = lgr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, lgr_pred))
```

```python
print(classification_report(y_test, lgr_pred))
```

#KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
error_rate = []

for i in range(1,40):
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train, y_train)
  pred_i = knn.predict(X_test)

  error_rate.append(np.mean(y_test != pred_i))
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(1,40), error_rate, marker='o', markerfacecolor='red')
plt.title("Error rate VS K")
plt.xlabel("K")
plt.ylabel("Error Rate")
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
kn_pred = knn.predict(X_test)
```

```python
print(confusion_matrix(y_test, kn_pred))
```

```python
print(classification_report(y_test, kn_pred))
```

#Descision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier(max_depth=15, min_samples_leaf=5, criterion='entropy')
```

#Random Forest
